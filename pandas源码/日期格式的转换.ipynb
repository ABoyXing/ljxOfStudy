```python
# 将各种日期字符串转成日期格式
import pandas as pd
df=pd.DataFrame({
    '原日期':['14-Feb-20','02/14/2020','2020.02.14','2020/02/14','20200214']
})
df['转换后的日期']=pd.to_datetime(df['原日期'])
df
```

```python
# 从多列中组合成一个日期
import pandas as pd
df=pd.DataFrame(
{
    'year':[2018,2019,2020],
    'month':[1,2,3],
    'day':[10,20,30],
    'hour':[13,14,15],
    'minute':[19,30,40],
    'second':[9,8,7]
})
df['组合后的日期']=pd.to_datetime(df)

df





```

```python
# dt对象的使用
import pandas as pd
df=pd.read_excel('原日期.xlsx')
df['日期']=pd.to_datetime(df['原日期'])
#df.info()
df['年']=df['日期'].dt.year  # 提取年

#提取月
df['月']=df['日期'].dt.month  

#提取日
df['日']=df['日期'].dt.day 

#提取星期
df['星期几']=df['日期'].dt.day_name()# 方法

#提取季度
df['季度']=df['日期'].dt.quarter

#判断
df['是否年底']=df['日期'].dt.is_year_end
df
```
