```python
#缺失值的处理方式 --删除
import pandas as pd
pd.set_option('display.unicode.east_asian.width',True)
df=pd.read_excel('msb课程记录.xls')
print(df)
print('----------------------------')
df=df.dropna()
print(df)
```

```python
import pandas as pd
pd.set_option('display.unicode.east_asian.width',True)
df=pd.read_excel('msb课程记录.xls')
print(df)
print('----------------------------')
df=df[df['课程总数量'].notnull()] # 提取课程数量中不为NaN 
print(df)
```

```python
#缺失值的处理填充
import pandas as pd
pd.set_option('display.unicode.east_asian.width',True)
df=pd.read_excel('msb课程记录.xls')
print(df)
print('----------------------------')
df['课程总数量']=df['课程总数量'].fillna(0)
print(df)
```
