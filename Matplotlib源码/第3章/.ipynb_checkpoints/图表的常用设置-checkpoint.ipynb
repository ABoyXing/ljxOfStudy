```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y,color='r',linestyle='-.') # 点

y=[10,15,20,25,30]
plt.plot(x,y,color='b',linestyle='--') # 双划线

y=[10,13,16,19,22]
plt.plot(x,y,color='g',linestyle=':')# 虚线

y=[10,18,26,34,42]
plt.plot(x,y,color='pink',linestyle='-')# 实线

plt.show()
```

```python
# 散点图
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,10,6,7,17]
plt.plot(x,y,'r*')
plt.show()
```

```python
# 绘制未来15天天气预报的折线图
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
df=pd.read_excel('未来15天天天气报.xlsx')
x=df['日期']
y=df['温度']
plt.plot(x,y,marker='o',color='r',mfc='w') #markerfacecolor简写mfc
plt.show()
```

```python

# 设置画布的颜色
# 绘制未来15天天气预报的折线图
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
df=pd.read_excel('未来15天天天气报.xlsx')

plt.figure(facecolor='yellow',figsize=(10,6))

x=df['日期']
y=df['温度']
plt.plot(x,y,marker='o',color='r',mfc='w') #markerfacecolor简写mfc
plt.show()
```

```python
# 设置坐标轴的标题
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
df=pd.read_excel('未来15天天天气报.xlsx')

# 用于处理中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.figure(figsize=(10,6))

x=df['日期']
y=df['温度']

plt.xlabel('2021年五一')  # x轴的标题
plt.ylabel('室外温度')    #y轴的标题
plt.plot(x,y,marker='o',color='r',mfc='w') #markerfacecolor简写mfc
plt.show()
```

```python
# 设置坐标轴的刻度
import matplotlib.pyplot as plt
import random
x=[i for i in range(1,11)] # 列表生成式
y=[random.randint(1,10) for _ in range(10)]

plt.plot(x,y,marker='o',mfc='w')

month=[str(i)+'月' for i in range(1,11)]

plt.xticks(range(1,11),month) # # 设置x轴坐标轴刻度
plt.yticks(range(1,11)) # 设置y轴坐标轴刻度

plt.xlim(1,10) # 设置坐标轴范围
plt.ylim(1,20)

plt.show()
```
