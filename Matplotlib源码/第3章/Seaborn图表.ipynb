```python
# 常用图表的绘制--折线图
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 

# 设置样式
sns.set_style('darkgrid')
# 解决中文乱码

plt.rcParams['font.sans-serif']=['SimHei']

# 读取数据
df=pd.read_excel('成绩表.xlsx')

sns.relplot(x='姓名',y='数学',kind='line',data=df)

plt.xticks(df['姓名'],rotation=45)

plt.show()
```

```python
# 线性回归图表分析，  小费
import matplotlib.pyplot as plt
import seaborn as sns


# 设置样式
sns.set_style('darkgrid')
# 解决中文乱码

plt.rcParams['font.sans-serif']=['SimHei']

# 读取数据
df=sns.load_dataset('tips') #关于小费数据集，是seaborn自带的

#绘制
sns.lmplot(x='total_bill',y='tip',data=df)

plt.show()
```

```python
# 绘制核密度图
import matplotlib.pyplot as plt
import seaborn as sns


# 设置样式
sns.set_style('darkgrid')
# 解决中文乱码

plt.rcParams['font.sans-serif']=['SimHei']

# 读取数据
df=sns.load_dataset('iris') #是seaborn自带的
#
sns.kdeplot(df['petal_width'],shade=True,color='r')

sns.kdeplot(df['petal_length'],shade=True,color='b')
plt.show()
```

```python
# 边际核密度图

import matplotlib.pyplot as plt
import seaborn as sns


# 设置样式
sns.set_style('darkgrid')
# 解决中文乱码

plt.rcParams['font.sans-serif']=['SimHei']

# 读取数据
df=sns.load_dataset('iris') #是seaborn自带的

# 绘制
sns.jointplot(x=df['sepal_length'],y=df['sepal_width'],kind='kde',space=0)

plt.show()
```
