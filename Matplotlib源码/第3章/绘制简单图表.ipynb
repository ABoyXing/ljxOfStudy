```python
# 绘制未来15天天气预报的折线图
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
df=pd.read_excel('未来15天天天气报.xlsx')
x=df['日期']
y=df['温度']
plt.plot(x,y)
plt.show()
```
